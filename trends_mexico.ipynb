```python
import feedparser
from datetime import datetime

fecha = datetime.now().strftime("%d/%m/%Y %H:%M")
url = "https://news.google.com/rss?hl=es-419&gl=MX&ceid=MX:es-419"

feed = feedparser.parse(url)

for noticia in feed.entries[:20]:
    print("📰", noticia.title)
```

```python
html = """
<h1>🔥 Tendencias México</h1>

"""

for noticia in feed.entries[:10]:

    html += f"""
    <p>Actualizado automáticamente: {fecha}</p>
    <div style="
        border:1px solid #ddd;
        padding:20px;
        margin-bottom:20px;
        border-radius:12px;
        box-shadow:0 2px 5px rgba(0,0,0,0.1);
    ">

        <h2 style="color:#222;">
            {noticia.title}
        </h2>

        <a 
            href="{noticia.link}" 
            target="_blank"
            style="
                background:#ff6600;
                color:white;
                padding:10px 15px;
                text-decoration:none;
                border-radius:8px;
                display:inline-block;
                margin-top:10px;
            "
        >
            Leer noticia
        </a>

    </div>
    """

print(html)
```

```python
with open("tendencias.html", "w", encoding="utf-8") as f:
    f.write(html)

print("HTML generado")
```
